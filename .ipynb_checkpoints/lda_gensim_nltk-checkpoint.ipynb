```python
#Prepare NLP and set stop words

nlp = spacy.load("en", disable=['ner'])

stop_words = list(spacy.lang.en.stop_words.STOP_WORDS)
    
stop_words.extend(['ai', 'new', 're', 'edu', 'use', 'not', 'would', 'say', 'could', 'be', 'know', 'good', 'go', 'get', 'do', 'done', 'try', 'many', 'some', 'nice', 'thank', 'think', 'see', 'rather', 'easy', 'easily', 'lot', 'lack', 'make', 'want', 'seem', 'run', 'need', 'even', 'right', 'line', 'even', 'also', 'may', 'take', 'come'])
stop_words.extend(['usa', 'american','french', 'france','china', 'italy' ,'italian', 'india', 'indian', 'mexico','mexican' 'japan' ,'denmark', 'danish', 'uk','uks','british', 'germany','german', 'sweden','london','edinburgh', 'cambridge'])
stop_words.extend(['strategy','strategic','federal','council','ministry','parliament','appendix'])
stop_words.extend(['villani', 'mathematician','author','woman'])
stop_words.extend(nltk.corpus.stopwords.words('english'))
```

```python
# Import Dataset into a Pandas Dataframe

def load_documents(eu_only=False):
    eu_list = ['denmark','france','germany','italy','sweden','uk']
    file_list = glob.glob(os.path.join(os.getcwd(), ".\\txts\\clean\\"+ "*.txt"))
            
    corpus = {}

    for file_path in file_list:
        filename = Path(Path(file_path).stem).stem
        if not eu_only or (eu_only and filename in eu_list):
            with open(file_path,encoding='utf-8',newline=None) as f_input:
                text = f_input.read().replace("\n", " ")
                clean_text = "".join([c for c in text if c.isprintable()])
                corpus[filename]= clean_text

    return corpus  

documents = load_documents(eu_only=False)
df = pd.DataFrame.from_dict(documents, orient='index', columns=['content'])
df.index.names = ['country']
df
```

```python
import re
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem.wordnet import WordNetLemmatizer

lmtzr = WordNetLemmatizer()
tokenizer = RegexpTokenizer(r'\w+')

def process_words_nltk(line):
    line = re.sub(r'https?:\/\/.*\s+', ' ', line, flags=re.MULTILINE) # remove URLs
    line = re.sub(r'\W+',' ',line, flags=re.MULTILINE) # remove non-word characters
    
    line = line.lower()
    tokens = tokenizer.tokenize(line)
    
    output = []
    for t in tokens:
        t = lmtzr.lemmatize(t)
        if (len(t) <= 2) or (len(t) >= 15) or (t in stop_words) or (t.isdigit()):
            continue
        output.append(t)
    return output



df['clean_content'] = df['content'].map(process_words_nltk)

# Add bigrams and trigrams to docs (only ones that appear 20 times or more).
common_terms = ["of", "with", "without", "and", "or", "the", "a" ,"as"]
bigram = gensim.models.Phrases(list(df['content']), common_terms=common_terms, min_count=10, threshold=20) # higher threshold fewer phrases.


def addNGrams(l):
    for token in bigram[l]:
        if '_' in token:
            l.append(token)
    return l

df['clean_content'] = df['clean_content'].map(addNGrams)

texts = list(df['clean_content'])

dictionary = corpora.Dictionary(texts)
#dictionary.filter_extremes(no_below=2, no_above=0.8)


N_TOPICS = 7 #or 7

# Create Corpus: Term Document Frequency
corpus = [dictionary.doc2bow(text) for text in texts]

# Build LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=dictionary,
                                           num_topics=N_TOPICS, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=1,
                                           passes=20,
                                           iterations=200,
                                           alpha='auto',
                                           per_word_topics=True)

pprint(lda_model.print_topics(num_words=30))
print('\nPerplexity: ', lda_model.log_perplexity(corpus))

# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, texts=texts, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)

```

##  What is the Dominant topic and its percentage contribution in each document

```python
def format_topics_sentences(ldamodel=None, corpus=corpus, texts=data):
    # Init output
    sent_topics_df = pd.DataFrame()

    # Get main topic in each document
    for i, row_list in enumerate(ldamodel[corpus]):
        row = row_list[0] if ldamodel.per_word_topics else row_list            
        # print(row)
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        # Get the Dominant topic, Perc Contribution and Keywords for each document
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # => dominant topic
                wp = ldamodel.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords] ,name=df.iloc[i].name))
            else:
                break
   

    # Add original text to the end of the output
    contents = pd.Series(texts,index=df.index)
    sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Keywords','Text']
    return(sent_topics_df)


df_topic_sents_keywords = format_topics_sentences(ldamodel=lda_model, corpus=corpus, texts=texts)

df_topic_sents_keywords
```

```python
# Get topic weights and dominant topics ------------
from sklearn.manifold import TSNE
from bokeh.plotting import figure, output_file, show
from bokeh.models import Label
from bokeh.io import output_notebook

# Get topic weights
topic_weights = []
for i, row_list in enumerate(lda_model[corpus]):
    topic_weights.append([w for i, w in row_list[0]])

# Array of topic weights    
arr = pd.DataFrame(topic_weights).fillna(0).values

# Keep the well separated points (optional)
arr = arr[np.amax(arr, axis=1) > 0.35]

# Dominant topic number in each doc
topic_num = np.argmax(arr, axis=1)

# tSNE Dimension Reduction
tsne_model = TSNE(n_components=N_TOPICS, verbose=1, random_state=0, angle=.99, init='pca')
tsne_lda = tsne_model.fit_transform(arr)

# Plot the Topic Clusters using Bokeh
output_notebook()
mycolors = np.array([color for name, color in mcolors.TABLEAU_COLORS.items()])
plot = figure(title="t-SNE Clustering of {} LDA Topics".format(N_TOPICS), plot_width=900, plot_height=700)
plot.scatter(x=tsne_lda[:,0], y=tsne_lda[:,1], color=mycolors[topic_num])
show(plot)
```
